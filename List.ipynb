```python
# Making of list
users =['val','bob','mfa','ron','ned']
print("List values               >> ",users)

# append to a list
users.append('mal')
print("List with Appended values >> ",users)

# Count of list
print("List Count                >> ",users.count('val'))

#Length of list
print("Length of list            >> ",len(users))

# list with multiple data types
multiList=['Bobby',100,20.84,'C']
print("List with Multiple data types >> ",multiList)

#insert values to a list
print("\nList BEFORE INSERT >>",users)
users.insert(2,'tran')
print("List AFTER INSERT >>",users)

#Extracting element from list thru Index
print("\nValue of Index 2  >>",users[2])
print("Value of Index 4  >>",users[4])
print("\nGetting last element of the list >>",users[-1])
print("Getting 2nd last element of the list >>",users[-2])

#Modifying the list elements
print("\nList value BEFORE modification >>",users)
users[0]='Bal'
users[-1]='yal'
print("List value AFTER modification >>",users)

# Removing elements from list
print("\nList value BEFORE removal  >>",users)

# removal with value
users.remove('mfa')
#removal with index
users.remove(users[2])
print('List value AFTER removal',users)

#POP the element - To work with particular value from the list
popUser=users.pop() # Pops the last element from the list
popUser
popFirst=users.pop(0) # Pops the First element from the list
popFirst
users
users.pop()
users

#Sorting the list
users.append('Aka')
users.append('ala')
users.append('zka')
users.append('pla')
users.append('ska')
users.append('rka')
print("\nList BEFORE SORT",users)
users.sort()
print("List AFTER sort",users)


#Reverse Sort
users.reverse()
print("Reverse sort the List",users)

# Sorted List & reverse thru parameter
sortedUserList = sorted(users,reverse=True)
print("Reverse Sort ",sortedUserList)

#Looping the list
for user in users:
    print("Wlecome "+user+" !!")

#List & Range
rangeList=list(range(1,25))
print("List creation with Range ",rangeList)

#List slicing
print("User List >> ",users)
firstThree=users[:3]
midFour=users[2:6]
lastThree=users[-3:]
print("First three>>",firstThree)
print("Middle four>>",midFour)
print("Last three>>",lastThree)

# Nested List
nestList = ["Bla",[5,7,8],['a']]
print(nestList)

#Copy a list
userCopy=users[:]
print(userCopy)

#Extend of the list

firstList =['h','i']
secondList=firstList.extend()
```
